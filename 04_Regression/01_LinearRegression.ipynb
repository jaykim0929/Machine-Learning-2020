# 단순 선형 회귀 (Linear Regression)

### 아버지와 아들의 키 - galton

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('http://www.randomservices.org/random/data/Galton.txt', sep='\t')
df.head()
```

```python
df.info()
```

```python
df = df[df.Gender == 'M']
height = df[['Father', 'Height']].rename(columns={'Height': 'Son'})
height.Father = height.Father * 2.54
height.Son = height.Son * 2.54
height.describe()
```

```python
plt.scatter(height.Father, height.Son)
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.grid()
plt.show()
```

### Numpy를 이용해서 회귀선 구하기

```python
A = np.vstack([height.Father, np.ones(len(height.Father))]).T
A.shape
```

```python
reg = np.linalg.lstsq(A, height.Son, rcond=None)
reg
```

```python
# 기울기와 절편
m, c =reg[0]
```

```python
# 잔차 제곱의 합
rss = reg[1][0]
```

```python
print(f'기울기: {m:.4f}, 절편{c:.4f}, 잔차제곱:{rss:.2f}')
```

```python
plt.figure(figsize=(8,7))
plt.scatter(height.Father, height.Son)
plt.plot([155, 200], [m*155+c, m*200+c], 'r', lw=3)
plt.title('The height of Father and Son')
plt.xlabel("Father's height (cm)")
plt.ylabel("Son's height (cm)")
plt.grid()
plt.show()
```

### Scikit-Learn을 이용해서 구하기

```python
from sklearn.linear_model import LinearRegression
```

```python
X = height.Father.values.reshape(-1,1)
y = height.Son.values
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
weight, bias = lr.coef_, lr.intercept_   # coefficient, intercept
rss = lr._residues
print(f'기울기: {weight[0]:.4f}, 절편: {bias:.4f}, 잔차제곱: {rss:.2f}')
```

```python
score = lr.score(X,y)
print(f'R_squared score: {score:.4f}')
```

### 당뇨병 데이터셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df.head()
```

```python
diabetes.target[:5]
```

### 당뇨병 vs BMI

```python
X = diabetes.data[:, 2].reshape(-1,1)
y = diabetes.target
```

```python
X.shape
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
weight, bias = lr.coef_, lr.intercept_  # coefficient, intercept
rss = lr._residues
print(f'기울기: {weight[0]:4f}, 절편: {bias:.4f}, 잔차제곱: {rss:.2f}')
```

### 훈련/테스트 데이터 셋 분리후 학습/예측/평가

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.05, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print(f'Mean Squared Error: {mse:.4f}, RMSE: {rmse:.4f}')
```

```python
score = r2_score(y_test, pred)
print(f'R_squared: {score:.4f}')
```

### 그래프로 확인

```python
plt.scatter(X_train, y_train)
plt.show()
```

```python
plt.scatter(X_train, y_train)
plt.plot([-0.1, 0.2], [-0.1*weight+bias, 0.2*weight+bias], 'r', lw=3)
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.1, 0.2], [-0.1*weight+bias, 0.2*weight+bias], 'r', lw=3)
plt.grid()
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print(f'Mean Squared Error: {mse:.4f}, RMSE: {rmse:.4f}')
score = r2_score(y_test, pred)
print(f'R_squared: {score:.4f}')
```

### 회귀 표본 만들기

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=100, n_features=1, n_targets=1, random_state=2021)
```

```python
plt.scatter(X, y)
plt.grid()
plt.show()
```

```python
X, y = make_regression(n_samples=100, n_features=1, n_targets=1, noise=100,         
                       random_state=2021)
plt.scatter(X, y)
plt.grid()
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X, y)
weight, bias = lr.coef_, lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot([-3, 4], [-3*weight+bias, 4*weight+bias], 'r', lw=3)
plt.grid()
plt.show()
```

```python
lr.score(X, y)
```
